
# calcular el numero de partidos jugados antes del inicio de cada competición

```python
def compute_win_streak(df):

    df=df.sort_values('date')
    df=df[["win"]]
    grouper = (df.win != df.win.shift()).cumsum()
    df['streak'] = df.groupby(grouper).cumsum()
    return df.streak.max(),df.iloc[-1].streak
    
```

```python
def compute_total_matches(df_stats,kickoff_day):
    df_stats=df_stats[df_stats.tournament!="FIFA World Cup"]
    data=[]
    for i in range (len(kickoff_day)):
        if i==0:
            df_statsD=df_stats[df_stats.date < kickoff_day[i] ]
        else:
            df_statsD=df_stats[(df_stats.date > kickoff_day[i-1])&(df_stats.date < kickoff_day[i])]
            
        for c in df_stats.country.unique():
            df_statsC=df_statsD[df_statsD.country==c]
            if not df_statsC.empty:
                max_streak,last_streak=compute_win_streak(df_statsC)
                data.append([kickoff_day[i],c,len(df_statsC),df_statsC.goles_marcados.sum(),df_statsC.goles_encajados.sum(),max_streak,last_streak])

    return pd.DataFrame(data, columns=['date','country','num_matches', 'goals scored','conceded goals',"max_win_streak","last_win_streak"])
        
        
```

```python
df_stats["year"]=df_stats.date.dt.year
df_statsWC=df_stats[df_stats.tournament=="FIFA World Cup"]
kickoff_day=df_statsWC.groupby("year").date.first().reset_index().date.values
df_statsbeforewrldcup=compute_total_matches(df_stats,kickoff_day)
```
